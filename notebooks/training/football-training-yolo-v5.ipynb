# Training Football Dataset by YOLO

Using a football dataset from Roboflow, we fine-tune the YOLO model to detect the ball, goalkeeper, players, and referee with higher accuracy and separate labeling.


##### Explaining Labels

```bash
2 0.21844791666666669 0.37982407407407404 0.02145312499999997 0.053870370370370325
```
Here `2` means object and rest are bounding boxes with x-centered, y-centered, width and height of the frame.

````python
import os 
base_dir = "~/Football-Analysis"
dataset_dir = os.path.join(base_dir, "training", "football-players-detection-1")
model_base_dir = os.path.join(base_dir, "models", "yolo26n.pt")
````

## Training the Data

````python
!yolo task=detect mode=train model=yolov5s.pt data={dataset_dir}/data.yaml epochs=100 imgsz=640

!yolo task=detect mode=train model=yolov5s.pt data={dataset_dir}/data.yaml epochs=100 imgsz=640 device=mps
````

````python
from ultralytics import YOLO
model = YOLO(model_base_dir)
results = model.train(
    data=f"{dataset_dir}/data.yaml",
    epochs=100,
    imgsz=640,
    device="mps",
)
````
